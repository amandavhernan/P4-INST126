```python
from tkinter import *
import numpy 
import pandas as pd

def showOutcome(gpa_entry, sat_entry, activities_entry, label):
    gpa = float(gpa_entry.get())
    sat = float(sat_entry.get())
    activities = int(activities_entry.get())
    rejection = True
    if (gpa >= 3.0 and sat >= 1250):
        rejection = False
        label.config(text="Congratulations!\n Our research shows that you have an excellent chance of being accepted into \n \
                          the University of Maryland.Maryland is one of the top public universities on the east coast,\n \
                          and it offers over 90 majors across 12 different colleges. The best major you could choose, though,\n \
                          is information science! And hereâ€™s why: collaborative classrooms environments, oppurtunities in data\n \
                          science, cybersecurity, digital curation, health, and tons of research opportunities!")
    # rejection
    if (gpa <=3.0 and sat < 1100):
        rejection = False
        label.config(text="Unfortunately, our research shows that you would have a low chance of being offered admission \n \
                           to the University of Maryland. Your best options would be to retake the SAT and get involved in \n \
                           more activities to boost your application.")
    # waitlisted
    for gpa in numpy.arange(2.7, 3.0):
        if sat in range(1100, 1250) and activities == 4:
            rejection = False
            label.config(text="Unfortunately, our research shows that you  will most likely  not be offered  admissions to the \n \
                               University of Maryland at this time. But since your application was strong, your appplication will \n \
                               most likely be placed in the waitlisted application pool. Your best options would be to retake the SAT \n \
                               and get involved in more activities to boost your application.")
    # deferred   
    for gpa in numpy.arange(2.7, 3.0):
        if sat in range(1100, 1250) and activities >= 5:
            rejection = False
            label.config(text="Unfortunately, our research shows you will most likely  not be offered  admissions to the \n \
                               University of Maryland at this time. But since your application was strong, your appplication \n \
                               will most likely be  deferred  to the next application pool. Your best options would be to retake \n \
                               the SAT and get involved in more activities to boost your application.")
    if rejection == True:
           label.config(text="Unfortunately, our research shows that you will not be offered admission to the \n \
                          University of Maryland. Your best options would be to retake the SAT and get involved \n \
                          in more activities to boost your application.")
        
def aboutInfo():
    win = Toplevel()
    Label(win, text="Welcome!\n \
                     This is a college admissions calculator that will determine your eligibility to attend the University of Maryland. \n \
                     Make sure you have your most up to date GPA and SAT score, as well as the number of activities you're currently involved \n \
                     in. This program will tell you if you will be accepted, rejected, waitlisted, or deferred based on your GPA, SAT, and the number of \n \
                     activities you're involved in.").grid(row=0)

def disclaimerInfo():
    win = Toplevel()
    Label(win, text="Please remember that the actual admissions process is based off much more than just numbers, \n \
                     so don't let this program's results discourage you from applying. The sole purpose of this program \n \
                     is to help you understand where you stand compared to other students.").grid(row=0)

master = Tk()
master.title("College Admissions Calculator")
master.geometry("1000x1000")
Label(master, text="GPA: ").grid(row=1)
Label(master, text="SAT SCORE: ").grid(row=2)
Label(master, text="ACTIVITIES: ").grid(row=3)

outcome_label = Label(master, text="")
outcome_label.grid(row=5, column = 0)

gpa_entry = Entry(master)
sat_entry = Entry(master)
activities_entry = Entry(master)

gpa_entry.grid(row=1, column=1)
sat_entry.grid(row=2, column=1)
activities_entry.grid(row=3, column=1)

about_button = Button(master, text="About", command=lambda: aboutInfo())
disclaimer_button = Button(master, text="Disclaimer", command=lambda: disclaimerInfo())
submit_button = Button(master, text="SUBMIT", command=lambda: showOutcome(gpa_entry, sat_entry, activities_entry, outcome_label))

about_button.grid(row=0, column=0)
disclaimer_button.grid(row=0, column=1)
submit_button.grid(row=4)
mainloop()
```
